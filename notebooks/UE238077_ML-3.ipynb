```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("house_price_data (1).csv")

x1 = df['Size (sq yd)'].values
x2 = df['Bedrooms'].values
x3 = df['Bathrooms'].values
x4 = df['Location (Sector)'].values
y  = df['Price (Lakhs)'].values

m = len(y)


def normalize(arr):
    mean = sum(arr) / len(arr)
    std = (sum((i - mean)**2 for i in arr) / len(arr))**0.5
    return [(i - mean) / std for i in arr]

x1 = normalize(x1)
x2 = normalize(x2)
x3 = normalize(x3)
x4 = normalize(x4)


theta0 = 0
theta1 = 0
theta2 = 0
theta3 = 0
theta4 = 0

alpha = 0.01
iterations = 500

cost_history = []


for it in range(iterations):

    sum_error = 0

    d0 = d1 = d2 = d3 = d4 = 0

    for i in range(m):

        y_pred = (theta0 +
                  theta1 * x1[i] +
                  theta2 * x2[i] +
                  theta3 * x3[i] +
                  theta4 * x4[i])

        error = y_pred - y[i]

        sum_error += error**2

        d0 += error
        d1 += error * x1[i]
        d2 += error * x2[i]
        d3 += error * x3[i]
        d4 += error * x4[i]

    theta0 = theta0 - alpha * (1/m) * d0
    theta1 = theta1 - alpha * (1/m) * d1
    theta2 = theta2 - alpha * (1/m) * d2
    theta3 = theta3 - alpha * (1/m) * d3
    theta4 = theta4 - alpha * (1/m) * d4

    cost = (1/(2*m)) * sum_error
    cost_history.append(cost)

    if it % 100 == 0:
        print("Iteration:", it, "Cost:", cost)


print("Final Thetas:")
print(theta0, theta1, theta2, theta3, theta4)


plt.plot(cost_history,'b.')
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Error vs Iteration")
plt.show()
```
